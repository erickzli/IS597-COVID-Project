```python
import covidcast
from datetime import date
```

```python
case = covidcast.signal("indicator-combination", "confirmed_incidence_num",
                        start_day=date(2020, 9, 1), end_day=date(2020, 9, 30),
                        geo_type="state", geo_values="il")

death = covidcast.signal("indicator-combination", "deaths_incidence_num",
                         start_day=date(2020, 9, 1), end_day=date(2020, 9, 30),
                         geo_type="state", geo_values="il")
```

```python
cases_v_deaths = covidcast.aggregate_signals([case, death])
cases_v_deaths = cases_v_deaths.rename(
    columns={"indicator-combination_confirmed_incidence_num_0_value": "cases",
              "indicator-combination_deaths_incidence_num_1_value": "deaths"})
cases_v_deaths[["time_value", "geo_value", "cases", "deaths"]].head(15)
```
