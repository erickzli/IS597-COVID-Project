## Data Notes
The employees_weekly_hours_US_2019_2020.csv file is from https://beta.bls.gov/dataViewer/view/timeseries/CES0500000002.
This dataset is all about the average weekly hours all US employees work in the private sector. The dataset is seasonally adjusted.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('employees_weekly_hours_US_2019_2020.csv')
```

```python
df
```

```python
df_2019 = df[df['Year'] == 2019]
#df_2019
```

```python
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_2019['Month'] = month
```

```python
df_2019
```

```python
df_2020 = df[df['Year'] == 2020]
#df_2020
```

```python
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_2020['Month'] = month
df_2020
```

```python
plt.plot(df_2020['Period'], df_2020['Value'], label = '2020')
plt.plot(df_2019['Period'], df_2019['Value'], label = '2019')
plt.xlabel('Month')
plt.ylabel('Average Weekly Hours')
plt.title('Average Weekly Hours of All Employees in the Private Sector')
plt.grid(True)
plt.legend()

#plt.savefig('IL_Pneumonia_Deaths_2019_2020.png', dpi=300, bbox_inches='tight')
plt.show()
```
